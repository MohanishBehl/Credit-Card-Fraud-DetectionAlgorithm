Importing the Dependencies

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading the dataset to a Pandas Dataframe
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
# first five rows of the dataset
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
# dataset information
credit_card_data.info()
```

```python
# checking the number pf missing values to each column
credit_card_data.isnull().sum()
```

```python
# distribution o legit transaction and fradulent transaction
credit_card_data['Class'].value_counts()
```

This Dataset is highly unbalanced

0 --> Noraml Trasaction
1 --> Fraudulent Transaction

```python
# separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both trasactions
credit_card_data.groupby('Class').mean()
```

Under-Sampling

Build a sarnple dataset containing similar distribution Of normal transactions ard Fraudulent Transactions.

Number Of Fraudulent Transactions --> 242

```python
legit_sample = legit.sample(n=242)
```

Concatenation two DataFrames

```python
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features & Targets

```python
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into Training data & Testing Data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression(max_iter=1000000, solver="lbfgs")
```

```python
# training he Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Store

```python
# accuracy on training store
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```
